```python
from sympy import *
from sympy import poly
x = symbols("x")

def reduction(expr, p) -> bool:
    p_exp = poly(expr)
    values = []
    for i in range(p):
        result = p_exp.subs('x', i) % p
        print(expr + ' with x:' + str(i) +' = ' + str(result))
        if values.__contains__(result): return False
        values.append(result)
    return True

#pass argument with format like '2*x^2+3*x+7'
def bijectivity(expr):
    is_reduction = reduction(expr, 4)
    if(is_reduction): print('polynomial is bijective')
    else: print('polynomial is not bijective')

#pass argument with format like '2*x^2+3*x+7'
def transitivity(expr) -> bool:
    is_reduction = reduction(expr, 8)
    if(is_reduction): print('polynomial is transitive')
    else: print('polynomial is not transitive')
```
